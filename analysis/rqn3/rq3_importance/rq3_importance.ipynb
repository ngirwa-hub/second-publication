```python
#concatnate two datasets
import pandas as pd

OUT_FILE = "merged_importance_rq3.csv"

# load
neu = pd.read_csv("importance_neutral_1256hrs.csv")
biased = pd.read_csv("importance_biased_llm1254hrs.csv")

# keep exactly these shared columns (in this order)
cols = [
    'source','row_id','base_model','variant_id','model','dc_solution',
    'rating','label','condition', 'bias_type', 'anchor_level','iteration','timestamp'
]

merged = pd.concat([neu[cols], biased[cols]], ignore_index=True)

merged.to_csv(OUT_FILE, index=False)
print("Saved:", OUT_FILE, "rows:", len(merged))
merged.info()
```
